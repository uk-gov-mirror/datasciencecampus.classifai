## Problem Overview

Statitical coding involves coding classifying data entries according to a scheme or taxonomy. 

How can historical coded samples be used to aid in coding new entries, either automatically or in an assisted manner?

![data problem](./images/data_problem_overview.png)

## ClassifAI Package - end to end solution

ClassifAI can be used to build systems that 'semantically' search historical coded dataset to find coded samples that are similar to uncoded samples.

```python
from classifai.vectorisers import HuggingfaceVectoriser
from classifa.indexers import VectorStore
from classifai.indexers.dataclasses import VectorStoreSearchInput

from classifai.indexers import VectorStore


vectoriser = HuggingfaceVectoriser(model_name = "all-MiniLM-L6-v2",)

vectorstore = VectorStore(
    file_name = "/path to data",
    data_type = "csv",
    vectoriser = vectoriser,
    output="vectorstore"
    )
```

```python
uncoded_input = VectorStoreSearchInput({'id':[1], 'query': ['organic cabbage farmer']})
uncoded_input
```

```python
vectorstore.search(uncoded_input)
```
